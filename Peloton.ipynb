```python
import pandas as pd
from datetime import date, datetime
import altair as alt
```

```python
peloton = pd.read_csv('Chasing_PRs_workouts.csv')

peloton = peloton[~peloton['Fitness Discipline'].str.contains('Meditation')]
peloton.sort_values(by=['Workout Timestamp'], inplace=True)
peloton['Workout Date'] = peloton['Workout Timestamp'].str.split(' ').str[0]

peloton_time = (datetime.strptime(peloton['Workout Date'].iloc[-1], '%Y-%m-%d'))-(datetime.strptime(peloton['Workout Date'][0], '%Y-%m-%d'))
peloton_time.days
print("You have worked out with Peloton over {} days.".format(peloton_time.days))
```

```python
peloton_instructors = peloton.groupby(['Instructor Name']).agg(['sum', 'count'])
peloton_instructors.sort_values(by=[('Type','count')], inplace=True, ascending=False)
print('You took the most classes with {}.'.format(peloton_instructors.index[0]))
peloton_instructors.sort_values(by=[('Length (minutes)','sum')], inplace=True, ascending=False)
print("You spend the most time with {}.".format(peloton_instructors.index[0]))
```

```python
peloton_prs = peloton[peloton['Total Output'].notna()]
peloton_prs = peloton_prs[peloton_prs['Fitness Discipline'] == 'Cycling']
peloton_prs = peloton_prs.sort_values(['Length (minutes)', 'Total Output'], ascending=[True, False])
peloton_prs
```

```python
line5 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#48C9B0')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=5))
)

line10 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#2471A3')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=10))
)

line20 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#76448A')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=20))
)

line30 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#922B21')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=30))
)

line45 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#2E4053')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=45))
)

line60 = alt.Chart(peloton_prs).mark_line().encode(
    alt.X('Workout Timestamp'),
    alt.Y('Total Output'),
    color=alt.value('#D4AC0D')
).transform_filter(
    (alt.FieldEqualPredicate(field='Length (minutes)', equal=60))
)

output_chart = line5 + line10 + line20 + line30 + line45 + line60
output_chart = output_chart.properties(
    title='Output over time'
).configure_title(
    fontSize=24,
    anchor='start'
)
output_chart
```
